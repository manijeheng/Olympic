```python
# ببرسی دیتا ست ورزشکارانی که در رده های مختلف ورزشی در ایونت های مختلف و سنین مختلف در سالهای مختلف توانسته اند در المبیک مدال بیاورند
#                  رسم نمودار در ژوبیتر   
# کتابخانه مورد نیاز :
# import numpy as np
# import matplotlib.pyplot as plt
# plt.plot(دیتاست)
# plt.plot(دیتاست,label='اسم نمودار')
# plt.xlabel('اسم محور افقی')
# plt.ylabel('اسم محور عمودی')
# plt.show()
# اشنایی با دستور count
# count() is used to count the number of non-NA/null observations across the given axis. 
# It works with non-floating type data as well.
# این دستور در سطر یا سنون  تعداد عیر تهی را برمیگرداند
# تابع count_value یک سری از مقادیر یونیک را برمیگرداند و برای هر کدام از این مقادیر یونیک میگوید چند بار تکرار شده  اند وو شامل مقادیر تهی نمیشود
# df['Id'].value_counts()
# این دستور براساس مثلا ستون ای دی مقادیر یونیک انرا با تعداد تکرار ان نمایش میدهد
import numpy as np
import pandas as pd
import seaborn as sns
from matplotlib import pyplot as plt

import os
print(os.listdir("../input"))

# data importing
data = pd.read_csv('../input/athlete_events.csv')
regions = pd.read_csv('../input/noc_regions.csv')

#  collect statistical information.Return the first 5 rows of the DataFrame.
data.head(5)
# Return a statistically description of the data in the DataFrame
data.describe()
# 
data.info()

regions.head(5)
# ادعام دو دیتاست بر اساس ستون مشترک  ان او سی ترکیب میشوندو 
# left: use only keys from left frame, similar to a SQL left outer join; preserve key order
merged = pd.merge(data, regions, on='NOC', how='left')

# Let's see the result:
merged.head()

# Let's start creating a new dataframe including only gold medalists.
# ساخت دیال فریمی از افرادی که مدال طلا گرفته اند
goldMedals = merged[(merged.Medal == 'Gold')]
goldMedals.head()

# بررسی مقدار تهی  در ستونها 
goldMedals.isnull().any()
# دستور ایز فینیت در ستون سن  عناصر تهی را تشخیص میدهد و انهای که مقادیر تهب ندارند را انتخاب میکند
goldMedals = goldMedals[np.isfinite(goldMedals['Age'])]

#  نمایش اماری و نموداری تعداد افرادی که در رده سنی مختلف مدال طلا گرفته اند
plt.figure(figsize=(20, 10))
plt.tight_layout()
sns.countplot(goldMedals['Age'])
plt.title('Distribution of Gold Medal
          
# چه تعداد نفر با سن بالا 50 مدال طلا دارند
          
goldMedals['ID'][goldMedals['Age'] > 50].count()

# ساخت دیتا فریمی از کسانی که بالا 50 سال سن دارند و ندال طلا دارند وبر اساس رشته ورزشی انها          
masterDisciplines = goldMedals['Sport'][goldMedals['Age'] > 50]
#  بیشترین افراد بالای 50 سال  با مدال طلامربوز  به تیر اندازان و اسب سواران است         
# figsize(float, float): These parameter are the width, height in inches.
plt.figure(figsize=(20, 10))
plt.tight_layout()
sns.countplot(masterDisciplines)
plt.title('Gold Medals for Athletes Over 50')  
          
# مطالعه دیتا بیس برا اساس زنان ورزشکار ندال اور
# ساخت دیتا بیسی برا اساس اینکه جنسیت زن و مربوط به فصل تابستان باشه
womenInOlympics = merged[(merged.Sex == 'F') & (merged.Season == 'Summer')]    
womenInOlympics.head(10)
          
# رسم نمودار و نمایش اماری ندال اورندگان زن در دهه های مختلف
          
sns.set(style="darkgrid")
plt.figure(figsize=(20, 10))
sns.countplot(x='Year', data=womenInOlympics)
plt.title('Women medals per edition of the Games')
          
# 1900ده مدال اورندگان دهه           
womenInOlympics.loc[womenInOlympics['Year'] == 1900].head(10)  
          
# شمارش تعداد مدال اورندگان زن بازیهای تابستانی دهه1990 بر اساس ای دی انها 
womenInOlympics['ID'].loc[womenInOlympics['Year'] == 1900].count() 
          
          
# 5 کشور اول از بیشترین مدال اورندگان          
#           دسته بندی یک دیتا ست بر اساس مثلا مقادیر ستون آ و تعداد تکرار مثلا عنصر ای ام تی را برای عر کدام از مقادیر آ وارد کنیم 
          
#           grouper = dftest.groupby('A')
#          df_grouped = grouper['Amt'].value_counts()
#           print (df_grouped.rename('count').reset_index())
goldMedals.region.value_counts().reset_index(name='Medal').head(5)
          
# رسم نمودار تعداد مدال ها برای هر مشور جزوه رده بنج کشور اول          
totalGoldMedals = goldMedals.region.value_counts().reset_index(name='Medal').head(5)
g = sns.catplot(x="index", y="Medal", data=totalGoldMedals,
                height=6, kind="bar", palette="muted")
g.despine(left=True)
g.set_xlabels("Top 5 countries")
g.set_ylabels("Number of Medals")
plt.title('Medals per Country') 
          
# حالا تحلیل اطلاعات ندال اورندگان اولین رده یعنی کشور امریکا
          
goldMedalsUSA = goldMedals.loc[goldMedals['NOC'] == 'USA']
          
# ساخت دیتا ستی شامل تعداد مدال های هر ایونت ورزشی برای کشور  امریکا و نمایش تنها 20 تایی اول این دیتا ست 
goldMedalsUSA.Event.value_counts().reset_index(name='Medal').head(20)
          
#  یافتن ورزشکاران مردی که در رشته بسکتبال شرکت داشنه اند و نمایش 15 تایی از این جستحو برا اساس ترتیب سال ان واقغه ورزشی         
basketballGoldUSA = goldMedalsUSA.loc[(goldMedalsUSA['Sport'] == 'Basketball') & (goldMedalsUSA['Sex'] == 'M')].sort_values(['Year'])
basketballGoldUSA.head(15)          
# اینبار ساخت دیتا فریمی از دیتا فریم قبلی که بر اساس سال آن واقغع ورزشی فقط اولین رکورد مربود به هر سال انها اورده شود          
groupedBasketUSA = basketballGoldUSA.groupby(['Year']).first()
groupedBasketUSA  
# شمارش تعداد سطر های این دیتا بیس
groupedBasketUSA['ID'].count()
          
# برررسی نسبت قد و وزن مدال اورندگان
# از انجایکه این ستون ها مقادیر تهی دارد
# ساخت دیتا بیسی که فقظ سطرهای با مقدار عیر تهی در آن باشد          
notNullMedals = goldMedals[(goldMedals['Height'].notnull()) & (goldMedals['Weight'].notnull())] 
# رسم نمودار براکندگی ارتباط قد و وزن و دیده میشه که یک ارتباط خطی بین قد ووزن وجود دارد
plt.figure(figsize=(12, 10))
ax = sns.scatterplot(x="Height", y="Weight", data=notNullMedals)
plt.title('Height vs Weight of Olympic Medalists')          
          
          
# مدال اورندگان با وزن بالا 160
notNullMedals.loc[notNullMedals['Weight'] > 160]     
          
#                      بررسی تعییرات تعداد زنان یا مردان شرکت کنند در المبیک با گدشت زمان    
# تقسیم دیتا فریم به دو بخش زنان و مردان بر اساس بازیهای فصل تابستانه فقط
MenOverTime = merged[(merged.Sex == 'M') & (merged.Season == 'Summer')]
WomenOverTime = merged[(merged.Sex == 'F') & (merged.Season == 'Summer')]
          
# دیتابیسی که شامل ورزشکاران ایتالیا باشد 
 itMenOverTime = MenOverTime.loc[MenOverTime['region'] == 'Italy']         
          
```
